```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import pandas as pd

bunga=datasets.load_iris()
print(dir(bunga))
print (bunga.feature_names)
print (bunga.target_names)
```

```python
sumbux=bunga.data[:,0]
sumbuy=bunga.data[:, 2]

plt.scatter(sumbux, sumbuy, c=bunga.target)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("Grafik Bunga Iris")
plt.show()
```

```python
#menggunakan model dbscan
dbscan=DBSCAN()

#fitting data
dbscan.fit(bunga.data)

#transformasi menggunakan PCA 2D
pca=PCA(n_components=2).fit(bunga.data)
pca_2d=pca.transform(bunga.data)


#visualisasi

#memberi label marker
label={0:'red', 1:'blue', 2:'green'}

for i in range (0, pca_2d.shape[0]):
    if dbscan.labels_[i]==0:
        cluster1=plt.scatter(pca_2d[i,0],
                            pca_2d[i,1],
                            c='r',
                            marker='^')
    elif dbscan.labels_[i]==1:
        cluster2=plt.scatter(pca_2d[i,0],
                            pca_2d[i,1],
                            c='g',
                            marker='.')
    elif dbscan.labels_[i]==-1:
        noise=plt.scatter(pca_2d[i,0],
                            pca_2d[i,1],
                            c='b',
                            marker='x')
#legend dan judul
plt.legend([cluster1, cluster2, noise],
          ['cluster1', 'cluster2', 'noise'])
plt.title('klasterisasi bunga irish dengan dbscan')

plt.show()
```
