```
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))

import IPython
import plotly as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)
import numpy as np
IPython.get_ipython().events.register('pre_run_cell', configure_plotly_browser_state)
```

```
x, y, z = np.random.multivariate_normal(np.array([0,0,0]),np.eye(3),200).transpose()
x.shape
```

```
trace = go.Scatter3d(
   x=x,
   y=y,
   z=z,
   mode='markers', 
   marker=dict(
       size=12,
       line=dict(
           color='rgba(217, 217, 217, 0.14)',
           width=0.5
       ),
       opacity=0.8
   )
)
data = [trace]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0,
    )
)
fig = go.Figure(data=data,layout=layout)
py.offline.iplot(fig)
```

```
import pandas as pd
url = 'https://raw.githubusercontent.com/plotly/datasets/master/iris.csv'
df = pd.read_csv(url)
df.head()
```

```
df.Name.unique()
```

```
x = df['SepalLength'][df['Name'] == 'Iris-virginica']
y = df['SepalWidth'][df['Name'] == 'Iris-virginica']
z = df['PetalLength'][df['Name'] == 'Iris-virginica']

trace1 = go.Scatter3d(
   x=x,
   y=y,
   z=z,
   mode='markers', 
   marker=dict(
       size=12,
       line=dict(
           color='rgba(217, 217, 217, 0.14)',
           width=0.5
       ),
       opacity=0.8
   )
)

x = df['SepalLength'][df['Name'] == 'Iris-versicolor']
y = df['SepalWidth'][df['Name'] == 'Iris-versicolor']
z = df['PetalLength'][df['Name'] == 'Iris-versicolor']

trace2 = go.Scatter3d(
   x=x,
   y=y,
   z=z,
   mode='markers', 
   marker=dict(
       size=12,
       line=dict(
           color='rgba(217, 217, 217, 0.14)',
           width=0.5
       ),
       opacity=0.8
   )
)

x = df['SepalLength'][df['Name'] == 'Iris-setosa']
y = df['SepalWidth'][df['Name'] == 'Iris-setosa']
z = df['PetalLength'][df['Name'] == 'Iris-setosa']

trace3 = go.Scatter3d(
   x=x,
   y=y,
   z=z,
   mode='markers', 
   marker=dict(
       size=12,
       line=dict(
           color='rgba(217, 217, 217, 0.14)',
           width=0.5
       ),
       opacity=0.8
   )
)
data = [trace1,trace2,trace3]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0,
    )
)
fig = go.Figure(data=data,layout=layout)
py.offline.iplot(fig)
```
