```python
kode=input(" Masukkan kode karyawan ")
nama=input(" Masukkan nama karyawan")
golongan=" "

while golongan !="A" and golongan !="B" and golongan !="C" and golongan !="D":
    golongan=input(" Masukkan Golongan :")
if golongan=="A":
    gaji=10000000.00
    potongan=0.025*gaji
elif golongan=="B":
    gaji=8500000.00
    potongan=0.02*gaji
elif golongan=="C":
    gaji=7000000.00
    potongan=0.015*gaji
elif golongan=="D":
    gaji=5000000.00
    potongan=0.01*gaji

status=0
while status !=1 and status !=2:
    status=int(input(" Masukkan status "))
if status==1:
    jumlahAnak=int(input(" masukkan jumlah anak "))
    tunjanganIstri=0.1*gaji
    tunjanganAnak=0.05*gaji*jumlahAnak
else:
    tunjanganIstri=0.0
    tunjanganAnak=0.0
    
gajiKotor=gaji+tunjanganIstri+tunjanganAnak

print("===================================================")
print(" STRUK GAJI KARYAWAN")
print("--------------------------------------------------")
print ("Nama Karyawan: ", nama, "( kode: ",kode,")")
print("golongan: ", golongan)
print("-------------------------------------------------")

print(" Gaji pokok :", gaji)
print("Potongan : ", potongan)
print("------------------------------------------------------------")
gajiBersih=gajiKotor-potongan
print("Gaji Bersih= ", gajiBersih)
```
